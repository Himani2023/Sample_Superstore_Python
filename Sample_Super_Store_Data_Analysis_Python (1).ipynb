```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
import plotly.io as pio
import plotly.colors as colors
pio.templates.default = "plotly_white"
```

```python
data = pd.read_csv("Sample - Superstore.csv",encoding = 'latin-1')
data.head(5)
```

```python
data.describe()
```

```python
data.info()
```

## ***Converting Date Columns***

```python
data["Order Date"] = pd.to_datetime(data['Order Date'])
data["Ship Date"] = pd.to_datetime(data['Ship Date'])
data.info()
```

```python
data["Order Month"] = data["Order Date"].dt.month_name()
data["Order Year"] = data["Order Date"].dt.year
data['Order day of weak'] = data["Order Date"].dt.dayofweek
data.head(2)
```

### ***Monthly Sales Analysis***

```python
Sales_by_month = data.groupby("Order Month")["Sales"].sum().reset_index()
Sales_by_month
```

```python
fig = px.line(Sales_by_month,x = "Order Month",y = "Sales",title = "Monthly Sales Analysis")
fig.show()
```

# ***Sales by Category***

```python
Sales_by_Category = data.groupby("Category")["Sales"].sum().reset_index()
Sales_by_Category
fig = px.pie(Sales_by_Category,
             values= "Sales",
             names = "Category",
             hole = 0.5,
             color_discrete_sequence=px.colors.qualitative.Pastel)
fig.update_traces(textposition = "inside",textinfo = "percent+label")
fig.update_layout(title_text= 'Sales Analysis by Category',title_font=dict(size=24))
fig.show()
```

# ***Sales Analysis by Sub- Category***

```python
Sales_by_Subcategory = data.groupby("Sub-Category")["Sales"].sum().reset_index()
fig = px.bar(Sales_by_Subcategory,
             x = "Sub-Category",
             y = "Sales",
             color = "Sub-Category",
             title = "Sales Analysis by Sub-Category")
fig.show()
```

# **Monthly Profit Analysis**

```python
Profit_by_month = data.groupby("Order Month")["Profit"].sum().reset_index()
fig = px.line(Profit_by_month,
             x = "Order Month",
             y = "Profit",
             title = "Monthly Profit Analysis")
fig.show()
```

# **Profit By Category**

```python
Profit_by_category = data.groupby("Category")["Profit"].sum().reset_index()
fig = px.pie(Profit_by_category,
             values= "Profit",
             names = "Category",
             hole = 0.5,
             color_discrete_sequence=px.colors.qualitative.Pastel)
fig.update_traces(textposition = "inside",textinfo = "percent+label")
fig.update_layout(title_text= 'Profit Analysis by Category',title_font=dict(size=24))
fig.show()

```

# ***Profit by Sub_Category***

```python
Profit_by_Subcategory = data.groupby("Sub-Category")["Profit"].sum().reset_index()
fig = px.bar(Profit_by_Subcategory,
             x = "Sub-Category",
             y = "Profit",
             title= "Profit Analysis by Sub-Category")

fig.show()
```

# *Sales & Profit- Customer Segment*

```python
Sales_profit_by_Segment = data.groupby("Segment").agg({'Sales':'sum','Profit':'sum'}).reset_index()
Sales_profit_by_Segment

color_palette = colors.qualitative.Pastel

fig = go.Figure()
fig.add_trace(go.Bar(x = Sales_profit_by_Segment["Segment"],
    y = Sales_profit_by_Segment["Sales"],
    name = "Sales",
    marker_color = color_palette[0]))

fig.add_trace(go.Bar(x = Sales_profit_by_Segment["Segment"],
    y = Sales_profit_by_Segment["Profit"],
    name = "Profit",
    marker_color = color_palette[1]))


fig.update_layout(title_text = "Sales & Profit by Customer Segment",
                  xaxis_title = " Customer Segment",
                  yaxis_title = "Amount")
fig.show()





```

# **Sales to Profit ratio**

```python
Sales_profit_by_Segment = data.groupby("Segment").agg({'Sales':'sum','Profit':'sum'}).reset_index()
Sales_profit_by_Segment["Sales_to_Profit_Ratio"] = Sales_profit_by_Segment["Sales"]/Sales_profit_by_Segment["Profit"]
print(Sales_profit_by_Segment[["Segment","Sales_to_Profit_Ratio"]])
```

### *Insignts & Recommnendations*
1. Technology is the most profitable category, especially sub-categories like Copiers and Phones.	Focus marketing and inventory investment on high-margin Technology items like Copiers and Accessories.

2. Furniture (Tables, Chairs, Bookcases) often results in losses or very low profit margins.	Reevaluate pricing, reduce discounts, or renegotiate supplier terms for Furniture sub-categories.

3. Office Supplies perform steadily but are not high in profitability.	Maintain current stock levels but explore bundling with Technology for upselling opportunities.

4. The Consumer segment generates highest revenue, but not always highest profit per order.	Target Corporate and Home Office segments for better profit margins with custom deals or subscriptions.

5. Central region shows underperformance compared to East and West in terms of profitability.	Launch region-specific promotions or investigate potential delivery/logistics challenges in Central.

6. Discounts often lead to negative profit, especially for low-margin items.	Use intelligent discounting  avoid blanket discounts and apply selectively based on product margins.

7. Standard shipping is used most often and is cost-efficient.	Encourage use of Standard Class for more orders. Offer faster shipping only when necessary.

8. High-profit sub-categories are sometimes underutilized (e.g., Copiers have high margin but low sales).	Increase visibility of profitable but less-sold items through better placement, bundling, or promotions.

9. Deep discounts on Furniture products cause huge losses.	Limit discounts on Furniture sub-categories unless necessary to clear excess stock.

10. Sales quantity does not always correlate with profit.	Focus on profit per unit rather than just volume. Adjust marketing goals accordingly.
